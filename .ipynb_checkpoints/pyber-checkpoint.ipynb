```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load and read city and ride data
city_data = pd.read_csv("Resources/city_data.csv")
ride_data = pd.read_csv("Resources/ride_data.csv")

# Combine the data into a single dataset
all_data = pd.merge(city_data, ride_data, on="city", how="outer")

# Display the data table for preview
all_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Urban Data
urbanDF = all_data.loc[all_data["type"] == "Urban", :]
urbanFares = urbanDF.groupby("city")["fare"].mean()
urbanRides = urbanDF.groupby("city")["ride_id"].count()
urbanDrivers = urbanDF.groupby("city")["driver_count"].mean()

# Suburban Data
suburbanDF = all_data.loc[all_data["type"] == "Suburban", :]
suburbanFares = suburbanDF.groupby("city")["fare"].mean()
suburbanRides = suburbanDF.groupby("city")["ride_id"].count()
suburbanDrivers = suburbanDF.groupby("city")["driver_count"].mean()

# Rural Data
ruralDF = all_data.loc[all_data["type"] == "Rural", :]
ruralFares = ruralDF.groupby("city")["fare"].mean()
ruralRides = ruralDF.groupby("city")["ride_id"].count()
ruralDrivers = ruralDF.groupby("city")["driver_count"].mean()

# Scatter Plots
plt.scatter(urbanRides, urbanFares, s=urbanDrivers*10, color="coral", edgecolor="black", alpha=.75, label="Urban", linewidth = .9)
plt.scatter(suburbanRides, suburbanFares, s=suburbanDrivers*10, color="skyblue", edgecolor="black", alpha=.75, label="Suburban", linewidth = .9)
plt.scatter(ruralRides, ruralFares, s=ruralDrivers*10, color="gold", edgecolor="black", alpha=.75, label="Rural", linewidth = .9)

# Graph Properties
plt.grid()
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fares ($)")

# Legend
lgnd = plt.legend(loc="best", title="City Types", fontsize="small")
lgnd.legendHandles[0]._sizes = [25]
lgnd.legendHandles[1]._sizes = [25]
lgnd.legendHandles[2]._sizes = [25]

# Text
plt.text(42, 35, "Note:\nCircle size correlates with driver count per city.")

# Save and Show
plt.savefig("BubblePlot.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Total Fares by City Type
urbanTotalFare = urbanDF["fare"].sum()
suburbanTotalFare = suburbanDF["fare"].sum()
ruralTotalFare = ruralDF["fare"].sum()

# Pie Chart
labels = ["Urban", "Suburban", "Rural"]
faresbycity = [urbanTotalFare, suburbanTotalFare, ruralTotalFare]
colors = ["coral", "skyblue", "gold"]
explode = (.1, 0, 0)
plt.pie(faresbycity, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", startangle=90, shadow=True)
plt.title("% of Total Fares by City Type")

# Save and Show
plt.savefig("FaresPie.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Total Rides by City Type
urbanTotalRides = urbanDF["ride_id"].count()
suburbanTotalRides = suburbanDF["ride_id"].count()
ruralTotalRides = ruralDF["ride_id"].count()

# Pie Chart
labels = ["Urban", "Suburban", "Rural"]
faresbycity = [urbanTotalRides, suburbanTotalRides, ruralTotalRides]
colors = ["coral", "skyblue", "gold"]
explode = (.1, 0, 0)
plt.pie(faresbycity, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", startangle=90, shadow=True)
plt.title("% of Total Rides by City Type")

# Save and Show
plt.savefig("RidesPie.png")
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Total Drivers by City Type
TotalDrivers = city_data.groupby(["type"])["driver_count"].sum() / city_data["driver_count"].sum()

# Pie Chart
labels = TotalDrivers.index
colors = ["gold", "skyblue", "coral"]
explode = (0, 0, .2)
plt.pie(TotalDrivers, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", startangle=0, shadow=True)
plt.title("% of Total Drivers by City Type")

# Save and Show
plt.savefig("DriversPie.png")
plt.show()
```

## Observable Trends

1. As city size increases (from rural to suburban to urban), there are increasingly more drivers and rides. This makes sense because the larger cities have larger populations both of drivers and of riders.

2. Though urban cities generate the largest total amount of fares, the average fare is highest in rural cities. Lower demand leads to higher costs in general.

3. City size appears to correlate with the total of drivers, fares, and riders. As city size increases, all three of these areas increase as well.
